**fillna method**

```
df.fillna(method='ffill')
```

```
df.fillna(method='pad')
```

```
df.fillna(method='bfill')
```

**fillna axis**

```
df.fillna(method='ffill',axis=0)
```

```
df.fillna(method='ffill',axis=1)
```

**df.limit**

```
df.fillna(0,limit=1)
```

```
df.fillna(method='ffill',limit=3)
```

**df.inplace**

```
df.fillna(method='ffill',inplace=True)
df
```
